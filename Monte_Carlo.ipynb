```python
#Import module and libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
```

```python
#define the function
def f(x):
    return np.sin(x)
```

```python
#check the shape of the function
x=np.linspace(0,1.5*np.pi,100)
y=f(x)
area = simps(y,x)
print(area)
```

```python
#plot the function
plt.plot(x,y)
```

```python
#Monte Carlo integration
width = 1.5*np.pi-0
samples = np.random.uniform(low = 0, high = width, size=10000)
monte_carlo_area = f(samples).mean()*width
print(monte_carlo_area)
```

```python
plt.plot(x, y, label="f(x)=$\sin(x)$")
plt.fill_between(x, 0, y, alpha=0.1)
plt.text(1, 0.75, f"Area from Simps is {area:0.3f}", fontsize=12)
plt.text(1, 0.5, f"Area from MC Integration is {monte_carlo_area:0.3f}", fontsize=12)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
```
