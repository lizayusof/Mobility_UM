```python
#Import module and libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
```

```python
#Declare your function
def fn(x):
    return x**3/(np.exp(x)-1)
```

```python
#Check the shape or behaviour of our plot
xs = np.linspace(0, 10, 100)
ys = fn(xs)
plt.plot(xs,fn(xs),label='Black body')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
```

```python
#Monte Carlo integration
width = 10-0
samples = np.random.uniform(low = 0, high = width, size=10000)
monte_carlo_area = f(samples).mean()*width
print(monte_carlo_area)
```

```python
plt.plot(xs, ys, label="Blackbody")
plt.fill_between(xs, 0, ys, alpha=0.1)
#plt.text(1, 0.75, f"Area from Simps is {area:0.3f}", fontsize=12)
plt.text(1, 0.5, f"Area from MC Integration is {monte_carlo_area:0.3f}", fontsize=12)
plt.xlabel("x")
plt.ylabel("f(x)")
```
